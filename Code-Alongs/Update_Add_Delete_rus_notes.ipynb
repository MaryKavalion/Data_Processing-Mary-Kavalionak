```python
import pandas as pd
```

```python
column_values = [f"Column{i}" for i in range (1,5)]
index_values = [chr(i) for i in range (65,72)]

df = pd.DataFrame([[f"{ind}{col}" for col in range (1,5)] for ind in index_values], index = index_values, columns = column_values)

df
```

## Update cells using loc or iloc

```python
copy = df.copy()

copy.iloc[1, 2] = "XX"
copy.iloc[0, :] = "YY"
copy.loc[["C", "E", "F"], ["Column2", "Column4"]] = "ZZ"
copy
```

```python
copy.iloc[:2, :2] = copy.iloc[-2:, -2:] + "XX"
copy
```

```python
copy.iloc[:, :] = 0
copy
```

## Adding row or columns

Rows and columns can be added using loc (but not iloc)

```python
copy = df.copy()
copy.loc["H"] = ["H1", "H2", "H3", "H4"]
copy.loc[:, "Column5"] = 0
copy.loc[:, "Column6"] = copy.loc[:, "Column1"] + "-" + copy.loc[:, "Column4"]
copy["Column7"] = copy["Column1"] + "-" + copy["Column6"]
copy.loc["I", ["Column3", "Column6"]] = 0
copy.loc["J", "new column"] = "Hello!"
copy
```

## Renaming columns and indexes

```python
copy = df.copy()
copy.columns = ["A", "B", "C", "D"]
copy.index = [0, 1, 2, 3, 4, 5, 6]
copy.rename(columns={"B": "Column B", "D": "Mary"}, inplace = True)
copy.rename(index={4: "Four"}, inplace = True)

copy
```

В коде выше, параметр inplace установлен в True, что означает, что изменения будут применены к самому объекту copy, и этот объект будет изменен непосредственно, а не создан новый объект с изменениями.

Если бы inplace был установлен в False (или не указан, так как значение по умолчанию - False), то метод rename создал бы новый DataFrame с измененными названиями столбцов, и надо было бы присвоить его переменной, чтобы сохранить изменения.

## Delete rows and columns

```python
copy = df.copy()
copy.drop(index=["B", "F"], columns = "Column3", inplace = True)

copy
```
