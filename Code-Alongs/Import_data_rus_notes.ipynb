header = 0: Этот аргумент указывает, что первая строка в файле CSV содержит заголовки столбцов (названия столбцов). Если header установлен в 0, то первая строка будет использована в качестве заголовков. 
Установка header = 1 в функции pd.read_csv() указывает Pandas на то, что вторая строка (нумерация строк начинается с 0) в файле CSV должна быть использована в качестве заголовков столбцов данных. То есть вторая строка файла будет интерпретирована как имена столбцов, и данные будут начинаться с третьей строки и далее.

У pd.read_csv() есть значение по умолчанию для аргумента header: "infer". Когда header="infer", Pandas попытается автоматически определить, есть ли заголовки в файле CSV. Если заголовки обнаруживаются в файле, они будут использованы в качестве названий столбцов; в противном случае, они будут сгенерированы числовыми индексами.

# Convert to common formats

```python
# Copy dataframe to clipboard (then use Ctrl+V)
heroes.to_clipboard()
```

```python
heroes.to_json("../TrainingData/new_heroes.json", orient = "records")
json_string = heroes.to_json(indent = 4, orient = "records")
print(json_string[:200])
```

Аргумент orient="records" указывает, что данные будут сохранены в формате JSON, где каждая строка DataFrame будет представлена как отдельный объект (запись) в JSON.

Параметр indent равен 4, что означает, что JSON объекты будут отформатированы с отступами в 4 пробела для лучшей читаемости.

print(json_string[:200]): Этот код выводит первые 200 символов из строки json_string в консоль.

orientstr

    Indication of expected JSON string format.

        Series:

                default is ‘index’

                allowed values are: {‘split’, ‘records’, ‘index’, ‘table’}.

        DataFrame:

                default is ‘columns’

                allowed values are: {‘split’, ‘records’, ‘index’, ‘columns’, ‘values’, ‘table’}.

        The format of the JSON string:

                ‘split’ : dict like {‘index’ -> [index], ‘columns’ -> [columns], ‘data’ -> [values]}

                ‘records’ : list like [{column -> value}, … , {column -> value}]

                ‘index’ : dict like {index -> {column -> value}}

                ‘columns’ : dict like {column -> {index -> value}}

                ‘values’ : just the values array

                ‘table’ : dict like {‘schema’: {schema}, ‘data’: {data}}


"split": Каждый столбец DataFrame сохраняется как отдельный JSON объект, и данные разбиваются на несколько JSON файлов.

"index": Используется для сохранения меток индекса DataFrame как ключей JSON объектов.

"columns": Используется для сохранения названий столбцов как ключей JSON объектов.

"values": Все значения DataFrame сохраняются в виде плоского списка, без сохранения структуры столбцов и индексов.

"table": Создается JSON объект с ключом "schema" и двумя ключами "fields" и "data". "fields" содержит информацию о схеме данных (названия столбцов и их типы), а "data" содержит сами данные.
